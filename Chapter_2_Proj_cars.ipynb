##### Note that preprocessing, done next, can involve several functions across multiple libraries. Functions that will apply to present features for feature engineering will be made alongside pipelines that address the corresponding feature(s). Also, a column transformer will be used to apply pipelines where appropriate and make_column_selector will select items of a specific data type. Note that FunctionTransformer is important when making your own functions.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import TransformedTargetRegressor, ColumnTransformer, make_column_selector
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
def column_two_product(X):
    return X[:, [0]] * X[:, [1]]

def two_product_name(function_transformer, feature_names_in):
    return ["product_of_two"]

def two_product_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_two_product, feature_names_out=two_product_name),
        StandardScaler())

def column_four_product(X):
    return X[:, [0]] * X[:, [1]] * X[:, [2]] * X[:, [3]]

def four_product_name(function_transformer, feature_names_in):
    return ["product_of_four"]

def four_product_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_four_product, feature_names_out=four_product_name),
        StandardScaler())

def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]

def ratio_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler())

log_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler())


minus_sqrt = make_pipeline(
    KNNImputer(),
    FunctionTransformer(lambda X: X**(-1/2), feature_names_out="one-to-one"),
    StandardScaler())

cbrt_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.cbrt, feature_names_out="one-to-one"),
    StandardScaler())

default_num_pipeline = make_pipeline(KNNImputer(), 
                                     StandardScaler())

default_cat_pipeline = make_pipeline(SimpleImputer(strategy="most_frequent"),
                                     OneHotEncoder(handle_unknown="ignore"))

preprocessing = ColumnTransformer([    
    ("Motor", two_product_pipeline(), ["Engine_size", "Power_perf_factor"]),
    ("Motor_Fuel", four_product_pipeline(), ["Engine_size", "Fuel_capacity", "Power_perf_factor", "Fuel_efficiency"]),
    ("Fuel_range", two_product_pipeline(), ["Fuel_capacity", "Fuel_efficiency"]),
    ("exp_minus_sqrt", minus_sqrt, ["Price_in_thousands"]),
    ("cbrt", cbrt_pipeline, ["Sales_in_thousands"]),
    ("log", log_pipeline, ["Engine_size", "Horsepower", "Wheelbase", "Fuel_capacity", "Fuel_efficiency", "Power_perf_factor"]),
    ("cat", default_cat_pipeline, make_column_selector(dtype_include=object)),
],
    remainder=default_num_pipeline)
```

#### Now the data can be put into models
##### Once a given dataset has been cleaned and prepared, it can be used to train models. We could also use metrics to verify the quality of the model's predictions. Note that this all applies to multiple algorithms, such as linear/ridge/lasso/elastic net regression models.

```python
from sklearn.linear_model import LinearRegression

lin_reg = make_pipeline(preprocessing, LinearRegression())
lin_reg.fit(cars, car_labels)
```

```python
car_predictions = lin_reg.predict(cars)
car_predictions[:10].reshape(-1, 1).round(2)
```

```python
np.array(car_labels[:10]).reshape(-1, 1)
```

##### It appears as though the linear regression model performs considerably well on the first 10 instances we use the predict method on. Still, we will consider alternative models and hyperparameter tuning in order to optimze our models and selection.

```python
from sklearn.metrics import mean_squared_error
lin_rmse = mean_squared_error(car_labels, car_predictions, squared=False)
lin_rmse
```

##### The root mean square error is *extremely* low for the training set. In order to check that this will remain consistent, we will use k-fold cross validation.

```python
from sklearn.model_selection import cross_val_score

lin_rmses = -cross_val_score(lin_reg, cars, car_labels, scoring="neg_root_mean_squared_error", cv=10)
```

```python
pd.Series(lin_rmses).describe()
```

##### While the mean squared error performance measure suggested this model was *phenomenal*, k-fold cross-validation suggests, in this case, that it is not *as* phenomenal, but still very good for this task based on the training set.

#### Fine-tune hyperparameters
##### A model's hyperparameters can make a huge difference in its performance, so it is important to optimize them.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

full_linear_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("linear", LinearRegression()),
])
```

```python
full_linear_pipeline.get_params().keys()
```

```python
param_grid = [
    {'preprocessing__log__knnimputer__n_neighbors':[1, 3, 7, 9],
     'preprocessing__exp_minus_sqrt__knnimputer__n_neighbors':[1, 5, 9, 11],
     'preprocessing__remainder__knnimputer__n_neighbors':[1, 5, 9, 13]},
]

grid_search = GridSearchCV(full_linear_pipeline, param_grid, cv=3, scoring="neg_root_mean_squared_error")
grid_search.fit(cars, car_labels)
```

```python
grid_search.best_params_
```

##### Note that depending on context, either grid searches or randomized searches will be preferred, as this choice can affect the computational and temporal expenses of a project.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {'preprocessing__log__knnimputer__n_neighbors':randint(1, 23),
                  'preprocessing__exp_minus_sqrt__knnimputer__n_neighbors':randint(1, 23),
                  'preprocessing__remainder__knnimputer__n_neighbors':randint(1, 23)}
lin_rnd_search = RandomizedSearchCV(
    full_linear_pipeline, param_distribs, n_iter=25, cv=3,
    scoring="neg_root_mean_squared_error", random_state=42)

lin_rnd_search.fit(cars, car_labels)    
```
