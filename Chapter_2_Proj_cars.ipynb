##### Note that preprocessing, done next, can involve several functions across multiple libraries. Functions that will apply to present features for feature engineering will be made alongside pipelines that address the corresponding feature(s). Also, a column transformer will be used to apply pipelines where appropriate and make_column_selector will select items of a specific data type. Note that FunctionTransformer is important when making your own functions.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import TransformedTargetRegressor, ColumnTransformer, make_column_selector
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
def column_two_product(X):
    return X[:, [0]] * X[:, [1]]

def two_product_name(function_transformer, feature_names_in):
    return ["product_of_two"]

def two_product_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_two_product, feature_names_out=two_product_name),
        StandardScaler())

def column_four_product(X):
    return X[:, [0]] * X[:, [1]] * X[:, [2]] * X[:, [3]]

def four_product_name(function_transformer, feature_names_in):
    return ["product_of_four"]

def four_product_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_four_product, feature_names_out=four_product_name),
        StandardScaler())

def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]

def ratio_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler())

log_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler())


minus_sqrt = make_pipeline(
    KNNImputer(),
    FunctionTransformer(lambda X: X**(-1/2), feature_names_out="one-to-one"),
    StandardScaler())

cbrt_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.cbrt, feature_names_out="one-to-one"),
    StandardScaler())

default_num_pipeline = make_pipeline(KNNImputer(), 
                                     StandardScaler())

default_cat_pipeline = make_pipeline(SimpleImputer(strategy="most_frequent"),
                                     OneHotEncoder(handle_unknown="ignore"))

preprocessing = ColumnTransformer([    
    ("Motor", two_product_pipeline(), ["Engine_size", "Power_perf_factor"]),
    ("Motor_Fuel", four_product_pipeline(), ["Engine_size", "Fuel_capacity", "Power_perf_factor", "Fuel_efficiency"]),
    ("Fuel_range", two_product_pipeline(), ["Fuel_capacity", "Fuel_efficiency"]),
    ("exp_minus_sqrt", minus_sqrt, ["Price_in_thousands"]),
    ("cbrt", cbrt_pipeline, ["Sales_in_thousands"]),
    ("log", log_pipeline, ["Engine_size", "Horsepower", "Wheelbase", "Fuel_capacity", "Fuel_efficiency", "Power_perf_factor"]),
    ("cat", default_cat_pipeline, make_column_selector(dtype_include=object)),
],
    remainder=default_num_pipeline)
```

### Now we can use cross-validation on candidate algorithms
##### Once a given dataset has been cleaned and prepared, it can be used to train models, but before this happens we should evaluate candidate algorithms using k-fold cross-validation. Afterwards, the selected candidate algorithm may use its fit() method on the entire training set to then be evaluated on the test set after tuning hyperparameters.

##### Multiple variations of linear regression will be evaluted, including the original linear regression algoritm, ridge regression, lasso regression, and elastic net regression. These variations have differing regularization techniques.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline

lin_reg = Pipeline([
    ("preprocessing", preprocessing), 
    ("lin_reg", LinearRegression())
])
lin_rmses = -cross_val_score(lin_reg, cars, car_labels, scoring="neg_root_mean_squared_error", cv=10)
pd.Series(lin_rmses).describe()
```

##### K-fold cross-validation on a linear regression model suggests no overfitting, so now we try other algorithms.

```python
from sklearn.linear_model import Ridge

ridge_reg = Pipeline([
    ("preprocessing", preprocessing),
    ("ridge_reg", Ridge(random_state=42))
])

ridge_rmses = -cross_val_score(ridge_reg, cars, car_labels, scoring="neg_root_mean_squared_error", cv=10)
pd.Series(ridge_rmses).describe()
```

##### The ridge regression algorithm seems to only perform slightly less well than linear regression. Note that after overviewing a few algorithms and selecting a few candidate models, we will fine-tune their hyperparameters.

```python
from sklearn.linear_model import Lasso

lasso_reg = Pipeline([
    ("preprocessing", preprocessing),
    ("lasso_reg", Lasso(random_state=42))
])

lasso_rmses = -cross_val_score(lasso_reg, cars, car_labels, scoring="neg_root_mean_squared_error", cv=10)
pd.Series(lasso_rmses).describe()
```

```python
from sklearn.linear_model import ElasticNet

elanet_reg = Pipeline([
    ("preprocessing", preprocessing),
    ("elasticnet_reg", ElasticNet(random_state=42))
])

elanet_rmses = -cross_val_score(elanet_reg, cars, car_labels, scoring="neg_root_mean_squared_error", cv=10)
pd.Series(elanet_rmses).describe()
```

### Hyperparameter Tuning

##### Note that the latter three algorithms each have hyperparameters that can be tuned, in addition to all four having preprocessing items that can also be tuned. We will tune some of these items in each of these estimators.

##### In order to see the items that can be tuned, we may use some methods and get extensive lists in this context based on the amount of preprocessing and hyperparameters the pipeline and algorithm have.

```python
lin_reg.get_params().keys()
```

##### Note that we can fine-tune a large amount of these items and will do so using a very manual method in the next cell. What hyperparameters and items are fine-tuned is a matter of choice and computational consideration.

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'preprocessing__log__knnimputer__n_neighbors':[1, 3, 5, 9],
     'preprocessing__remainder__knnimputer__n_neighbors':[1, 3, 7, 9, 11]},
]

grid_search = GridSearchCV(lin_reg, param_grid, cv=5, scoring="neg_root_mean_squared_error")
grid_search.fit(cars, car_labels)
```

```python
grid_search.best_params_
```

##### In spite of having used k-fold cross-validation and hyperparameter tuning, we have not yet trained the linear regression model on any of the training set and will do so using the best_params_ of a randomized search.

##### A randomized search works in a similar manner to the grid search, except that it should be used for large hyperparameter spaces instead of a smaller list of possible options and it can go on for a selected count of iterations. 

##### The randint() function from the scipy library's stats package will be used to randomly select the integers for the hyperparameter tuning. Due to there being randomness, a random_state hyperparameter is set for reproducibility.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_grid = {'preprocessing__log__knnimputer__n_neighbors':randint(1, 23),
              'preprocessing__remainder__knnimputer__n_neighbors':randint(1, 23)}


rnd_search = RandomizedSearchCV(lin_reg, param_distributions=param_grid, n_iter=25, 
                                 cv=5, scoring="neg_root_mean_squared_error", random_state=42)
rnd_search.fit(cars, car_labels)
```

```python
rnd_search.best_params_
```
