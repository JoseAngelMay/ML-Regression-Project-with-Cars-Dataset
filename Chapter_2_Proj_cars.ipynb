```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import TransformedTargetRegressor, ColumnTransformer, make_column_selector
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
def column_product(X):
    return X[:, [0]] * X[:, [1]]

def product_name(function_transformer, feature_names_in):
    return ["product"]

def product_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_product, feature_names_out=product_name),
        StandardScaler())

def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]

def ratio_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler())

log_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler())


minus_sqrt = make_pipeline(
    KNNImputer(),
    FunctionTransformer(lambda X: X**(-1/2), feature_names_out="one-to-one"),
    StandardScaler())

cbrt_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.cbrt, feature_names_out="one-to-one"),
    StandardScaler())

default_num_pipeline = make_pipeline(KNNImputer(), 
                                     StandardScaler())

default_cat_pipeline = make_pipeline(SimpleImputer(strategy="most_frequent"),
                                     OneHotEncoder(handle_unknown="ignore"))

preprocessing = ColumnTransformer([    
    ('Engine_Curb_Ratio', ratio_pipeline(), ["Engine_size", "Curb_weight"]),
    ("Fuel_range", product_pipeline(), ["Fuel_capacity", "Fuel_efficiency"]),
    ("Wheelbase_Length_Ratio", ratio_pipeline(), ["Wheelbase", "Length"]),
    ("Area", product_pipeline(), ["Width", "Length"]),
    ("exp_minus_sqrt", minus_sqrt, ["Price_in_thousands"]),
    ("cbrt", cbrt_pipeline, ["Sales_in_thousands"]),
    ("log", log_pipeline, ["Engine_size", "Horsepower", "Wheelbase", "Fuel_capacity", 
                           "Fuel_efficiency", "Power_perf_factor"]),
    ("cat", default_cat_pipeline, make_column_selector(dtype_include=object)),
],
    remainder=default_num_pipeline)
```

```python
prepared_cars = preprocessing.fit_transform(cars)
```

#### Now the data can be put into models

#### Fine-tune hyperparameters
##### A model's hyperparameters can make a huge difference in its performance, so it is important to optimize them.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

full_ridge_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("ridge", Ridge(alpha=0.1)),
])
```

```python
full_ridge_pipeline.get_params().keys()
```

```python
param_grid = [
    {'preprocessing__remainder__knnimputer__n_neighbors':[1, 3, 9],
     'preprocessing__log__knnimputer__n_neighbors': [1, 3, 13],
     'ridge__alpha': [0.3, 0.7, 1]},
]

grid_search = GridSearchCV(full_ridge_pipeline, param_grid, cv=3, scoring="neg_root_mean_squared_error")
grid_search.fit(cars, car_labels)
```
